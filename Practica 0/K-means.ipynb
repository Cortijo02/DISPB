Práctica 2: Algoritmo K-Means
El siguiente código crea dos nubes de puntos con forma de media luna.
A continuación aplica el algoritmo K-Means.

```python
import numpy as np
from sklearn.cluster import KMeans
#from scikit-learn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
#from scikit-learn import make_moons

# Genera datos de ejemplo
np.random.seed(0)
#X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0], [4,3]])

# Generar un array de 10000 elementos [x, y]
#X = np.random.rand(10000, 2)

X, _ = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1], c='r')
plt.show()
# Crea un objeto KMeans
kmeans = KMeans(n_clusters=2, random_state=0)

# Aplica el algoritmo
kmeans.fit(X)

# Obtiene los centroides y las etiquetas de los clústeres
centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_

# Visualiza los resultados
plt.scatter(X[:, 0], X[:, 1], c=etiquetas, cmap='viridis')
plt.scatter(centroides[:, 0], centroides[:, 1], marker='X', s=200, color='red')
plt.show()
```

Ejercicio 1:
Modificar los parámetros de K-Means para poder clasificar los dos conjuntos de datos empleando las siguientes configuraciones para su generación:
Noise = 0.05
Noise = 0.1
Noise = 0.15
Noise = 0.13

```python
noises = [.05, .1, .15, .13]
num_clusters = [7, 7, 7, 7]

fig, axs = plt.subplots(2, 4, figsize=(15, 7))

for i, (noise, num_cluster) in enumerate(zip(noises, num_clusters)):
    X, _ = make_moons(n_samples=200, noise=noise, random_state=0)
    axs[0, i].scatter(X[:, 0], X[:, 1], c='r')
    axs[0, i].set_title(f"Original data ({noise=})")
    axs[0, i].set_xticks([])
    axs[0, i].set_yticks([])

    kmeans = KMeans(n_clusters=num_cluster, random_state=33)

    kmeans.fit_predict(X)

    centroides = kmeans.cluster_centers_
    etiquetas = kmeans.labels_

    axs[1, i].scatter(X[:, 0], X[:, 1], c=etiquetas, cmap='viridis')
    axs[1, i].scatter(centroides[:, 0], centroides[:, 1], marker='X', s=200, color='red')
    axs[1, i].set_title(f"K-means ({noise=})")
    axs[1, i].set_xticks([])
    axs[1, i].set_yticks([])

plt.show()
```
