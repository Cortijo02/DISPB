Práctica 1: Algoritmo DBSCAN
El siguiente código crea dos nubes de puntos con forma de media luna.
A continuación aplica el algoritmo DBSCAN con una distancia entre clústeres de 0.3 y un número mínimo de muestras para formar un cluster de 5.

```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Genera datos de ejemplo (en forma de "luna")
X, _ = make_moons(n_samples=200, noise=0.05, random_state=0)
print("Conjunto de datos original")
plt.scatter(X[:, 0], X[:, 1], c='r')
plt.title("Original data")
plt.show()

# Crea un objeto DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)

# Aplica el algoritmo
y_dbscan = dbscan.fit_predict(X)

# Visualiza los resultados
print("Conjunto de datos clasificados")
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("DBSCAN Clustering")
plt.show()
```

Modificar los parámetros de DBSCAN para poder clasificar los dos conjuntos de datos empleando las siguientes configuraciones para su generación:
Noise = 0.05
Noise = 0.1
Noise = 0.15
Noise = 0.13

```python
noises = [.05, .1, .15, .13]
epsilons = [.3, .28, .25, .25]
min_samples = [5, 8, 8, 8]

fig, axs = plt.subplots(2, 4, figsize=(15, 7))

for i, (noise, eps, min_samples) in enumerate(zip(noises, epsilons, min_samples)):
    X, _ = make_moons(n_samples=200, noise=noise, random_state=0)
    axs[0, i].scatter(X[:, 0], X[:, 1], c='r')
    axs[0, i].set_title(f"Original data ({noise=})")
    axs[0, i].set_xticks([])
    axs[0, i].set_yticks([])

    dbscan = DBSCAN(eps=eps, min_samples=min_samples)

    y_dbscan = dbscan.fit_predict(X)

    axs[1, i].scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis')
    axs[1, i].set_title(f"DBSCAN ({noise=})")
    axs[1, i].set_xticks([])
    axs[1, i].set_yticks([])

plt.show()
```
